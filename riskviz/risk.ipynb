```python
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

```python
data = pd.DataFrame()
pop = 1000
chance = 60
bacon = int(60 * 0.18)
none = pop - chance - bacon

data['None'] = [none]
data['Chance'] = [chance]
data['Bacon Eater'] = [bacon]

print(data)

data.plot.bar()
data.plot.bar(stacked=True)

data.plot.barh()
data.plot.barh(stacked=True)

data.T.plot.pie(subplots=True)
```

```python
print(data['None'].values)
a1 = [0]*data['None'].values[0]
a2 = [1]*data['Chance'].values[0]
a3 = [2]*data['Bacon Eater'].values[0]

a1.extend(a2)
a1.extend(a3)
#random.shuffle(a1)
```

```python
b = np.array(a1).reshape((20,50))
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(b, linewidths=0.5, cbar=False,yticklabels=False,xticklabels=False, cmap='Blues',center=0.5)
```

```python
random.shuffle(a1)
b2 = np.array(a1).reshape((20,50))
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(b2, linewidths=0.5, cbar=False,yticklabels=False,xticklabels=False, cmap='Blues',center=0.5)
```
