```python
import pandas as pd
import matplotlib.pyplot as plt

path = "images/"

# Read the data
weather = pd.read_csv('data/london2018.csv')

# Plot some charts and save them as images
weather.plot(x='Month', y = 'Tmax', linewidth=5)
plt.savefig(f'{path}temp1.png')

weather.plot.bar(x='Month', y = 'Rain')
plt.savefig(f'{path}temp2.png')
```

```python

fig, ax = plt.subplots(figsize=(15,10), layout = 'constrained')


img1 = plt.imread(f'{path}temp1.png')
img2 = plt.imread(f'{path}temp2.png')
bground = plt.imread(f'{path}pexels-brett-sayles-1431822-pale.jpg')


title = """
The Weather:

    London, 2018
"""
descr = """
You can see from the line chart
that it was warm in the summer and 
much colder in the winter.

The bar chart shows us that it
hardly rained in July.

This is probably not a great surprise.
"""

# Create a figure that we can use as a canvas
ax.plot([0,100],[0,100], alpha = 0)


# Place the images on the canvas
ax.imshow(img1, aspect='auto', alpha=1, extent=(50,100,50,100))
ax.imshow(img2, aspect='auto', alpha=1, extent=(0,50,0,50))
ax.imshow(bground, aspect='auto', alpha=0.5, extent=(0,100,0,100))

# Add text to the canvas
ax.text(5,95, title,
        verticalalignment='top', horizontalalignment='left',
        color='red', fontsize=36, weight = 'bold')
ax.text(55,45, descr,
        verticalalignment='top', horizontalalignment='left',
        color='black', fontsize=18)

# Switch off the ticks and bounding frame
plt.axis('off')


plt.savefig(f'weather-info-graphic.png')
```
