```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame({'category':['Sales','Service','Expenses','Taxes','Interest'],
                   'num':[100,10,-20,-30,60]})
```

```python
df['tot'] = df['num'].cumsum()
df['tot1']= df['tot'].shift(1).fillna(0)
df
```

```python
fig,ax = plt.subplots()
ax.bar(x=df['category'],height=df['num'],bottom= df['tot1'])
ax.bar(x=df['category'], height=df['tot1'], alpha = 0.1)
```

```python
df.loc[df['num'] >= 0, 'color'] = 'green'
df.loc[df['num'] < 0, 'color'] = 'red'
df
```

```python
fig,ax = plt.subplots()
ax.bar(x=df['category'],height=df['num'],bottom= df['tot1'], color=df['color'])
ax.bar(x=df['category'], height=df['tot1'], alpha = 0.1)
```

```python

fig,ax = plt.subplots()
ax.bar(x=df['category'],height=df['num'],bottom= df['tot1'], color=df['color'])
ax.bar(x=df['category'], height=df['tot1'], alpha = 0.1)
for i, v in enumerate(df['num']):
        plt.text(x = i, y =df['tot1'][i]+df['num'][i]/2 ,s=f"{df['num'][i]}")

```
