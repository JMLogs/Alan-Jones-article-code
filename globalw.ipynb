# Topical Plots: Global Warming
## What chart will get the message across best

```
import pandas as pd
gw = pd.read_csv('https://data.giss.nasa.gov/gistemp/graphs/graph_data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data/graph.txt',skiprows=(0,1,2,4), delimiter=' ', skipinitialspace=True)
gw.columns
```

```
import matplotlib
matplotlib.style.use('seaborn')
```

```
gw=gw.rename(columns={'No_Smoothing':'Temp','Lowess(5)':"Smoothed"})
gw
```

```
gw.plot(x='Year',figsize=(20,5))
```

```
ax=gw.plot(x='Year',y='Smoothed')
gw.plot.scatter(x='Year',y='Temp',c='Temp', colormap='inferno',figsize=(20,5),ax=ax)
```

```
import seaborn as sns
```

```
gw2=gw.drop(columns='Temp')
gw3=gw.drop(columns='Smoothed')
gw2
```

```
gw2=gw2.set_index('Year')
sns.set(rc={'figure.figsize':(20,5)})
sns.heatmap(gw2.T)
gw2
```

```
gw3=gw3.set_index('Year')
sns.set(rc={'figure.figsize':(20,5)})
sns.heatmap(gw3.T)
gw3
```

```
url='https://raw.githubusercontent.com/alanjones2/dataviz/master/tempbydecade.csv'
gwdec = pd.read_csv(url, delimiter=' ', skipinitialspace=True)
gwdec=gwdec.set_index('Year')
gwdec
```

```
sns.set(rc={'figure.figsize':(20,5)})
sns.heatmap(gwdec)
```

```
gw4=gw3.sort_values('Temp',ascending=False)
sns.set(rc={'figure.figsize':(5,30)})
sns.heatmap(gw4)
gw4[:20]
```

```
gw4=gw3.sort_values('Temp',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.heatmap(gw4[:20])
```
